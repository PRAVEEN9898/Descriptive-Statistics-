1. How many people are there in each department?

```python
df["Department"].value_counts()
```

```python
a = plt.bar(x=df["Department"].value_counts().index,height=df["Department"].value_counts())
plt.bar_label(a)
plt.ylabel("NO:OF People")
plt.xlabel("Name of Departments")
plt.show()
```

```python
sns.barplot(x=df["Department"].value_counts().index,y=df["Department"].value_counts())
plt.show()
```

```python
plt.figure(figsize=(10,5))
a = plt.bar(x=df["Department"].value_counts().index,height=df["Department"].value_counts(),label="Departments")
b = plt.bar(x=df["Department"].value_counts().index,height=df["Gender"].value_counts(),label="Gender")
plt.xticks(df["Gender"].unique())
plt.xticks(df["Department"].unique())
plt.legend()
plt.bar_label(a)
plt.bar_label(b)
plt.show()
```
